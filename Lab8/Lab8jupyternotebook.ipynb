```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
ccdb=pd.read_csv("UCI_Credit_Card.csv")

ccdb.shape
```

```python
#nice way to look at some of data
ccdb.describe().transpose()
```

```python
x = ccdb.drop("default.payment.next.month", axis=1)
y = ccdb["default.payment.next.month"]
```

```python
#split data into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=1693)

#transform and fit the data
scale_x = StandardScaler()
x_train = scale_x.fit_transform(x_train)
x_test = scale_x. transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix

#three hidden layers of 13 nodes each
mlp = MLPClassifier(hidden_layer_sizes=(13,13,13), max_iter=500)
```

```python
#fit the neural network to the training data, and use the fitted network to make a prediction
mlp.fit(x_train, y_train)
pred = mlp.predict(x_test)
#print(pred)
```

```python
#this is a prediction of default (1) or not defaulting (0)
pred
```

```python
nn_cm = confusion_matrix(y_test, pred)
#print(nn_cm)
#nn_cm
```

```python
print_cm(nn_cm, ["No Default", "Default"])
```
